```python


data = pd.DataFrame({
    'Season': season,
    'Temperature': tem_cal
})


data['Season'] = pd.to_datetime(
    data['Season'],
    format='%m/%Y'
)

temp = np.array(data['Temperature'])


if len(temp) % 2 != 0:
    temp = temp[:-1]


half = len(temp) // 2

first_half = temp[:half]
second_half = temp[half:]


x = np.sort(first_half)
y = np.sort(second_half)


n = len(x)

low_end = int(n / 3)
mid_end = int(2 * n / 3)

x_low = x[:low_end]
y_low = y[:low_end]

x_mid = x[low_end:mid_end]
y_mid = y[low_end:mid_end]

x_high = x[mid_end:]
y_high = y[mid_end:]

low_slope = (np.mean(y_low) - np.mean(x_low)) / np.mean(x_low)

mid_slope = (np.mean(y_mid) - np.mean(x_mid)) / np.mean(x_mid)

high_slope = (np.mean(y_high) - np.mean(x_high)) / np.mean(x_high)


print(f"Low Temperature ITA Slope    : {low_slope:.4f}")
print(f"Medium Temperature ITA Slope : {mid_slope:.4f}")
print(f"High Temperature ITA Slope   : {high_slope:.4f}")


def interpret_trend(value):

    if value > 0:
        return "Increasing Trend"

    elif value < 0:
        return "Decreasing Trend"

    else:
        return "No Trend"

print("Low Temperature Trend    :", interpret_trend(low_slope))
print("Medium Temperature Trend :", interpret_trend(mid_slope))
print("High Temperature Trend   :", interpret_trend(high_slope))


plt.figure(figsize=(10,10))


plt.scatter(
    x_low,
    y_low,
    s=70,
    alpha=0.8,
    label='Low Temperature'
)

plt.scatter(
    x_mid,
    y_mid,
    s=70,
    alpha=0.8,
    label='Medium Temperature'
)

plt.scatter(
    x_high,
    y_high,
    s=70,
    alpha=0.8,
    label='High Temperature'
)


minimum = min(min(x), min(y))
maximum = max(max(x), max(y))

plt.plot(
    [minimum, maximum],
    [minimum, maximum],
    linewidth=3,
    linestyle='--',
    label='1:1 Line'
)


plt.xlabel(
    'First Half Sorted Temperature (°C)',
    fontsize=14
)

plt.ylabel(
    'Second Half Sorted Temperature (°C)',
    fontsize=14
)

plt.title(
    'Advanced Innovative Trend Analysis (ITA)\n'
    'Calabar Temperature Classification',
    fontsize=16,
    fontweight='bold'
)


plt.grid(True)

plt.legend(fontsize=12)


plt.savefig(
    'Advanced_ITA_Calabar.png',
    dpi=600,
    bbox_inches='tight'
)


plt.show()
```

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import rankdata


data = pd.DataFrame({
    'Season': season,
    'Temperature': tem_cal
})


data[['Month', 'Year']] = data['Season'].str.split('/', expand=True)

data['Year'] = data['Year'].astype(int)

annual_temp = data.groupby('Year')['Temperature'].mean().reset_index()

years = annual_temp['Year'].values
temp = annual_temp['Temperature'].values


def pettitt_test(x):

    n = len(x)

    ranks = rankdata(x)

    U = np.zeros(n)

    for t in range(n):

        U[t] = 2 * np.sum(ranks[:t+1]) - (t+1) * (n+1)

    K = np.max(np.abs(U))

    change_point = np.argmax(np.abs(U))

    p_value = 2 * np.exp((-6 * K**2) / (n**3 + n**2))

    return change_point, K, p_value, U

cp, K, p, U = pettitt_test(temp)

cp_year = years[cp]



print(f"Change Point Year : {cp_year}")
print(f"Test Statistic K  : {K:.2f}")
print(f"P-value           : {p:.5f}")


if p < 0.05:
    print("\nSignificant change point detected.")
else:
    print("\nNo significant change point detected.")

plt.figure(figsize=(15,7))

plt.plot(
    years,
    temp,
    marker='o',
    linewidth=2,
    label='Annual Mean Temperature'
)

plt.axvline(
    cp_year,
    color='red',
    linestyle='--',
    linewidth=3,
    label=f'Change Point ({cp_year})'
)

plt.title(
    'Pettitt Change Point Detection\nCalabar Temperature Series',
    fontsize=18,
    fontweight='bold'
)

plt.xlabel('Year')
plt.ylabel('Temperature (°C)')

plt.legend()

plt.grid(True)

plt.savefig(
    'pettitt_temperature_change_point.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()

plt.figure(figsize=(15,6))

plt.plot(
    years,
    U,
    linewidth=2,
    label='Pettitt U Statistic'
)

plt.axvline(
    cp_year,
    color='red',
    linestyle='--',
    linewidth=3,
    label=f'Change Point ({cp_year})'
)

plt.title(
    'Pettitt Statistic Series',
    fontsize=16,
    fontweight='bold'
)

plt.xlabel('Year')
plt.ylabel('U Statistic')

plt.legend()

plt.grid(True)

plt.savefig(
    'pettitt_U_statistic.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()


annual_temp.to_csv(
    'annual_temperature_series.csv',
    index=False
)

```

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    r2_score
)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Dense,
    LSTM,
    Conv1D,
    MaxPooling1D
)

from tensorflow.keras.callbacks import EarlyStopping


df = pd.DataFrame({
    'Date': pd.to_datetime(season, format='%m/%Y'),
    'Temperature': tem_cal
})

df.set_index('Date', inplace=True)

data = df[['Temperature']].values

scaler = MinMaxScaler(feature_range=(0,1))

scaled_data = scaler.fit_transform(data)

train_size = int(len(scaled_data) * 0.80)

train_data = scaled_data[:train_size]
test_data  = scaled_data[train_size:]

time_step = 12

def create_dataset(dataset, time_step=12):

    X, y = [], []

    for i in range(len(dataset) - time_step):

        X.append(dataset[i:(i + time_step), 0])
        y.append(dataset[i + time_step, 0])

    return np.array(X), np.array(y)

X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test   = create_dataset(test_data, time_step)


X_train = X_train.reshape(
    X_train.shape[0],
    X_train.shape[1],
    1
)

X_test = X_test.reshape(
    X_test.shape[0],
    X_test.shape[1],
    1
)

model = Sequential()

model.add(
    Conv1D(
        filters=64,
        kernel_size=2,
        activation='relu',
        input_shape=(time_step,1)
    )
)

model.add(MaxPooling1D(pool_size=2))

model.add(LSTM(100, activation='relu'))

model.add(Dense(50, activation='relu'))

model.add(Dense(1))

model.compile(
    optimizer='adam',
    loss='mse'
)

model.summary()

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

history = model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size=16,
    verbose=1,
    callbacks=[early_stop]
)

train_predict = model.predict(X_train)
test_predict  = model.predict(X_test)

train_predict = scaler.inverse_transform(train_predict)
test_predict  = scaler.inverse_transform(test_predict)

y_train_actual = scaler.inverse_transform(
    y_train.reshape(-1,1)
)

y_test_actual = scaler.inverse_transform(
    y_test.reshape(-1,1)
)

rmse = np.sqrt(
    mean_squared_error(y_test_actual, test_predict)
)

mae = mean_absolute_error(
    y_test_actual,
    test_predict
)

mse = mean_squared_error(
    y_test_actual,
    test_predict
)

r2 = r2_score(
    y_test_actual,
    test_predict
)


mape = np.mean(
    np.abs(
        (y_test_actual - test_predict) / y_test_actual
    )
) * 100

nse = 1 - (
    np.sum((y_test_actual - test_predict)**2)
    /
    np.sum(
        (y_test_actual - np.mean(y_test_actual))**2
    )
)

print(f"RMSE : {rmse:.4f}")
print(f"MAE  : {mae:.4f}")
print(f"MSE  : {mse:.4f}")
print(f"R²   : {r2:.4f}")
print(f"MAPE : {mape:.4f}%")
print(f"NSE  : {nse:.4f}")

future_dates = pd.date_range(
    start='2026-02-01',
    end='2035-12-01',
    freq='MS'
)

future_steps = len(future_dates)

temp_input = scaled_data[-time_step:].reshape(
    1,
    time_step,
    1
)

future_predictions = []

for i in range(future_steps):

    pred = model.predict(temp_input, verbose=0)

    future_predictions.append(pred[0,0])

    temp_input = np.append(
        temp_input[:,1:,:],
        [[[pred[0,0]]]],
        axis=1
    )

future_predictions = np.array(
    future_predictions
).reshape(-1,1)

future_predictions = scaler.inverse_transform(
    future_predictions
)

train_plot = np.empty_like(data)

train_plot[:] = np.nan

train_plot[
    time_step:len(train_predict)+time_step
] = train_predict

test_plot = np.empty_like(data)

test_plot[:] = np.nan

test_plot[
    len(train_predict)+(time_step*2):len(data)
] = test_predict

plt.figure(figsize=(22,10))

plt.plot(
    df.index,
    data,
    color='black',
    linewidth=2.5,
    marker='o',
    markersize=3,
    label='Observed Trend'
)

plt.plot(
    df.index,
    train_plot,
    color='blue',
    linewidth=2,
    marker='s',
    markersize=3,
    label='Train Prediction'
)

plt.plot(
    df.index,
    test_plot,
    color='orange',
    linewidth=2,
    marker='^',
    markersize=4,
    label='Test Prediction'
)

plt.plot(
    future_dates,
    future_predictions,
    color='red',
    linewidth=3,
    linestyle='--',
    marker='D',
    markersize=4,
    label='Future Prediction (2026-2035)'
)

xticks = pd.date_range(
    start='1981-01-01',
    end='2035-01-01',
    freq='YS'
)

plt.xticks(
    xticks,
    [d.strftime('%m/%Y') for d in xticks],
    rotation=45
)

plt.xlabel(
    'Date',
    fontsize=14,
    fontweight='bold'
)

plt.ylabel(
    'Temperature (°C)',
    fontsize=14,
    fontweight='bold'
)

plt.title(
    'CNN-LSTM Hybrid Model for Temperature Prediction (1981-2035)',
    fontsize=18,
    fontweight='bold'
)

plt.grid(
    True,
    linestyle='--',
    alpha=0.4
)

plt.legend(
    fontsize=12,
    loc='upper left'
)

metrics_text = (
    f"RMSE  = {rmse:.3f}\n"
    f"MAE   = {mae:.3f}\n"
    f"MSE   = {mse:.3f}\n"
    f"R²    = {r2:.3f}\n"
    f"MAPE  = {mape:.3f}%\n"
    f"NSE   = {nse:.3f}"
)

plt.text(
    0.22,
    0.80,
    metrics_text,
    transform=plt.gca().transAxes,
    fontsize=12,
    bbox=dict(
        facecolor='white',
        edgecolor='black',
        alpha=0.9
    )
)

plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

tem_cal = np.array(tem_cal)


scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(tem_cal.reshape(-1,1))

def create_dataset(data, time_step=12):  
    X, y = [], []
    for i in range(len(data) - time_step):
        X.append(data[i:i+time_step, 0])
        y.append(data[i+time_step, 0])
    return np.array(X), np.array(y)

time_step = 12
X, y = create_dataset(data_scaled, time_step)

train_size = int(len(X) * 0.8)

X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

model = Sequential()
model.add(Dense(64, activation='relu', input_dim=time_step))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')

model.fit(X_train, y_train,
          epochs=100,
          batch_size=8,
          verbose=1)

train_pred = model.predict(X_train)
test_pred  = model.predict(X_test)


train_pred = scaler.inverse_transform(train_pred)
test_pred  = scaler.inverse_transform(test_pred)

y_train_actual = scaler.inverse_transform(y_train.reshape(-1,1))
y_test_actual  = scaler.inverse_transform(y_test.reshape(-1,1))

rmse = np.sqrt(mean_squared_error(y_test_actual, test_pred))
mae  = mean_absolute_error(y_test_actual, test_pred)
r2   = r2_score(y_test_actual, test_pred)
mape = np.mean(np.abs((y_test_actual - test_pred) / y_test_actual)) * 100

nse = 1 - (np.sum((y_test_actual - test_pred)**2) /
           np.sum((y_test_actual - np.mean(y_test_actual))**2))

print(f"RMSE : {rmse:.4f}")
print(f"MAE  : {mae:.4f}")
print(f"R²   : {r2:.4f}")
print(f"MAPE : {mape:.2f}%")
print(f"NSE  : {nse:.4f}")


future_steps = 120  

last_input = data_scaled[-time_step:].flatten()
future_predictions = []

for _ in range(future_steps):
    pred = model.predict(last_input.reshape(1, -1))
    future_predictions.append(pred[0,0])
    
    last_input = np.append(last_input[1:], pred)

future_predictions = scaler.inverse_transform(
    np.array(future_predictions).reshape(-1,1)
)

years_numeric = season

future_labels = []
last_month, last_year = map(int, years_numeric[-1].split('/'))

for _ in range(future_steps):
    last_month += 1
    if last_month > 12:
        last_month = 1
        last_year += 1
    future_labels.append(f"{last_month}/{last_year}")

train_plot = np.full(len(tem_cal), np.nan)
test_plot  = np.full(len(tem_cal), np.nan)

train_plot[time_step:time_step+len(train_pred)] = train_pred.flatten()
test_plot[time_step+len(train_pred):time_step+len(train_pred)+len(test_pred)] = test_pred.flatten()


plt.figure(figsize=(20,8))

# Observed
plt.plot(season, tem_cal, label='Observed')

# Train/Test
plt.plot(season, train_plot, '--', label='Train Predicted')
plt.plot(season, test_plot, '--', label='Test Predicted')

# Future
all_labels = season + future_labels

plt.plot(range(len(season), len(all_labels)),
         future_predictions,
         '--', marker='o', label='Future Forecast')


step = 12
plt.xticks(ticks=range(0, len(all_labels), step),
           labels=all_labels[::step],
           rotation=45)

plt.title("ANN Temperature Prediction (Calabar): 1981–2035")
plt.xlabel("Time (Month/Year)")
plt.ylabel("Temperature (°C)")

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
